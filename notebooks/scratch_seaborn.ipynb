```python
import seaborn as sns
%matplotlib inline 
#to plot the graphs inline on jupyter notebook
```

```python
t=sns.load_dataset('tips')
#to check some rows to get a idea of the data present
t.head()
```

```python

sns.barplot(x='day',y='total_bill',data=t)
```

```python
sns.countplot(x='sex',data=t)
```

```python
sns.boxplot(x='day',y='total_bill',data=t,palette='rainbow')
```

```python

sns.boxplot(x="day",y="total_bill",hue="smoker",data=t)
```

```python
sns.violinplot(x="day",y="total_bill",data=t,hue='sex',palette='Set1')

```

```python

sns.stripplot(x="day",y="total_bill",data=t,jitter=True,hue='sex',palette='Set1')
```
