
# Now we try to find any common elements based on the common column of `SK_ID_CURR`


```python

set_application_data = set(application_data[['SK_ID_CURR']].to_dict()['SK_ID_CURR'].values())

set_previous_application = set(previous_application[['SK_ID_CURR']].to_dict()['SK_ID_CURR'].values())

len(set_application_data.intersection(set_previous_application))
```

```python

list_of_common_loan_ids = list(set_application_data.intersection(set_previous_application))
```
